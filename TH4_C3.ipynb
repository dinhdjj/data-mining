```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/RestaurantDataset.csv', header=None, names=['area', 'type', 'rating'])

df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df
```

```python
areas = df['area'].unique()
for area in areas:
    df['area_'+area] = df['area'].apply(lambda x: 'Yes' if x == area else '')

types = df['type'].unique()
for type in types:
    df['type_'+type] = df['type'].apply(lambda x: 'Yes' if x == type else '')

ratings = df['rating'].unique()
for rating in ratings:
    df['rating_'+rating] = df['rating'].apply(lambda x: 'Yes' if x == rating else '')

df
```

```python
columns = list(set(df.columns) - set(['area', 'type', 'rating']))

df.loc[:,columns].to_csv('datasets/normalized_restaurant.csv', index=False)
```
