```python
import pandas as pd
import regex as re
```

## Import data

```python
df = pd.read_csv('datasets/car.csv', delimiter='|', error_bad_lines=False)

df
```

## Normalize "km" filed

```python
WEIGHT = 1

def number_km(km):
    km = re.findall(r'[\d,]+', km)
    km = re.sub(r'[,]+', '', km[0])
    return int(km)

df['number_km'] = df['km'].apply(number_km)

max_km = df['number_km'].max()
min_km = df['number_km'].min()

def normalize_km(km):
    return (km - min_km) / (max_km - min_km) * WEIGHT

df['normalized_km'] = df['number_km'].apply(normalize_km)
```

## Normalize "imp_exp" filed

```python
WEIGHT = 1

def normalize_imp_exp(imp_exp):
    result = 0
    if imp_exp == 'Nhập khẩu':
        result = 1
    return result * WEIGHT

df['normalized_imp_exp'] = df['imp_exp'].apply(normalize_imp_exp)
```

## Normalize "car_type" filed

```python
WEIGHT = 1

types = {
    'SUV': 1,
    'Coupe': 0.9,
    'Sedan': 0.8,
    'Crossover': 0.7,
    'Convertible/Cabriolet': 0.6,
    'Hatchback': 0.5,
    'Van/Minivan': 0.4,
    'Bán tải / Pickup': 0.3,
    'Truck': 0.2,
    'Wagon': 0,
}

def normalize_car_type(car_type):
    return types[car_type] * WEIGHT

df['normalized_car_type'] = df['car_type'].apply(normalize_car_type)
```

## Normalize "out_color" filed

```python
WEIGHT = 1

types = {
    'Đen': 1,
    'Nâu': 1,
    'Tím': 0.8,
    'Đồng': 0.7,
    'Đỏ': 0.6,
    '-': 0.5,
    'Nhiều màu': 0.5,
    'Màu khác': 0.5,
    'Hồng': 0.4,
    'Kem': 0.4,
    'Cam': 0.3,
    'Cát': 0.3,
    'Vàng': 0.3,
    'Ghi': 0.2,
    'Xanh': 0.1,
    'Xám': 0.1,
    'Bạc': 0,
    'Trắng': 0,
}

def normalize_out_color(out_color):
    return types[out_color] * WEIGHT

df['normalized_out_color'] = df['out_color'].apply(normalize_out_color)
```

## Normalize "in_color" filed

```python
WEIGHT = 1

types = {
    'Đen': 1,
    'Nâu': 1,
    'Tím': 0.8,
    'Đồng': 0.7,
    'Đỏ': 0.6,
    '-': 0.5,
    'Nhiều màu': 0.5,
    'Màu khác': 0.5,
    'Hồng': 0.4,
    'Kem': 0.4,
    'Cam': 0.3,
    'Cát': 0.3,
    'Vàng': 0.3,
    'Ghi': 0.2,
    'Xanh': 0.1,
    'Xám': 0.1,
    'Bạc': 0,
    'Trắng': 0,
}

def normalize_in_color(in_color):
    return types[in_color] * WEIGHT

df['normalized_in_color'] = df['in_color'].apply(normalize_in_color)
```

## Normalize "door_num" filed

```python
WEIGHT = 1

def number_door_num(door_num):
    door_num = re.findall(r'[\d]+', door_num)
    return int(door_num[0])

df['number_door_num'] = df['door_num'].apply(number_door_num)

max_num = df['number_door_num'].max()
min_num = df['number_door_num'].min()

def normalize_door_num(door_num):
    return (door_num - min_num) / (max_num - min_num) * WEIGHT

df['normalized_door_num'] = df['number_door_num'].apply(normalize_door_num)
```

## Normalize "seat_num" filed

```python
WEIGHT = 1

def number_seat_num(seat_num):
    seat_num = re.findall(r'[\d]+', seat_num)
    return int(seat_num[0])

df['number_seat_num'] = df['seat_num'].apply(number_seat_num)

max_num = df['number_seat_num'].max()
min_num = df['number_seat_num'].min()

def normalize_seat_num(seat_num):
    return (seat_num - min_num) / (max_num - min_num) * WEIGHT

df['normalized_seat_num'] = df['number_seat_num'].apply(normalize_seat_num)
```

## Normalize "new_old" filed

```python
WEIGHT = 1

def normalize_new_old(new_old):
    result = 0
    if new_old == 'Xe cũ':
        result =  1
    return result * WEIGHT

df['normalized_new_old'] = df['new_old'].apply(normalize_new_old)
```

## Normalize "car_year" filed

```python
WEIGHT = 1

def number_car_year(car_year):
    car_year = re.findall(r'[\d]+', car_year)
    return int(car_year[0])

df['number_car_year'] = df['car_year'].apply(number_car_year)

max_num = df['number_car_year'].max()
min_num = df['number_car_year'].min()

def normalize_car_year(car_year):
    return (car_year - min_num) / (max_num - min_num) * WEIGHT

df['normalized_car_year'] = df['number_car_year'].apply(normalize_car_year)
```

## Normalize "price" filed

```python
WEIGHT = 1

def number_price(price):
    price = re.findall(r'[\d]+', price)
    if len(price) == 2:
        return int(price[0]) * 1000 + int(price[1])
    return int(price[0])

df['number_price'] = df['price'].apply(number_price)

max_num = df['number_price'].max()
min_num = df['number_price'].min()

def normalize_price(price):
    return (price - min_num) / (max_num - min_num) * WEIGHT

df['normalized_price'] = df['number_price'].apply(normalize_price)
```

## Save to arff file

```python
import arff

columns = [
    'normalized_km',
    'normalized_imp_exp',
    'normalized_car_type',
    'normalized_out_color',
    'normalized_in_color',
    'normalized_door_num',
    'normalized_seat_num',
    'normalized_new_old',
    'normalized_car_year',
    'normalized_price',
]
df = df.loc[:, columns]

arff.dump('datasets/normalized_car.arff', df.values, names=columns)

df
```
