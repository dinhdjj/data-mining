```python
import pandas as pd
```

```python
from math import nan

df = pd.read_csv('datasets/Full_Mark_2020.csv', low_memory=False)
df = df.loc[
    (df['Ngoaingu'].notnull()) 
    & (df['Toan'].notnull()) 
    & (df['Nguvan'].notnull())
    & ~(df['KHTN'].isnull() & df['KHXH'].isnull())
]

def mon1(row):
    if(row['KHTN'] != nan):
        return row['Hoahoc']
    return row['GDCD']

def mon2(row):
    if(row['KHTN'] != nan):
        return row['Vatli']
    return row['Diali']

def mon3(row):
    if(row['KHTN'] != nan):
        return row['Sinhhoc']
    return row['LichSu']

df['Mon1'] =  df.apply(mon1, axis=1);
df['Mon2'] =  df.apply(mon2, axis=1);
df['Mon3'] =  df.apply(mon3, axis=1);

df
```
