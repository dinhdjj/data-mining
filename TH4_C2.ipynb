```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/Hcm_Mark_2020.csv')

df
```

```python
columns = ['Toan', 'Nguvan', 'Ngoaingu', 'Hoahoc', 'Sinhhoc', 'Vatli']
df = df.loc[df['KHTN'].notnull(), columns]
```

```python
def normalize(diem):
    if diem > 8:
        return 'Yes'
    return ''

for column in columns:
    df['normalized_' + column ] = df[column].apply(normalize)

df = df.drop(columns=columns)
```

```python


df.to_csv('datasets/Hcm_Mark_2020_TH4.csv', index=False)
```
